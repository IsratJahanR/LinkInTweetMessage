```python
import pandas as pd
import numpy as np
df =pd.read_csv("dataset.csv")
df
```

```python
domain_count=pd.read_csv('Domain_count.csv')
domain_count
```

```python
domain_count.drop(['Unnamed: 0',],axis=1,inplace=True)
domain_count
```

```python
df = pd.merge(df, domain_count, on='Domain')
df=df.sort_values(by='Unnamed: 0')
df          
```

```python
def categorize(value):
    if value > 20:
        return 'common'
    elif 2 <= value < 21:
        return 'sometimes'
    else:
        return 'rare'
```

```python
# Add a new column 'Status' based on the conditions
df['strata'] = df['Count'].apply(categorize)
df
```

```python
df.to_csv('tweet_link_domain_strata.csv')
df
```
