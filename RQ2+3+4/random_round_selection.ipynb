```python
import pandas as pd
df = pd.read_csv('tweet_link_domain_strata.csv')
df
```

```python
df.drop(['Unnamed: 0.1',],axis=1,inplace=True)
df
```

```python
strt = df.groupby('strata').size().reset_index(name='Count')
#year = year.sort_values(by='Count',ascending=False)
strt
```

```python
rare = df[(df['strata'] =='rare')]
sometimes = df[(df['strata'] =='sometimes')]
common = df[(df['strata'] =='common')]
print(len(rare)," ",len(sometimes)," ",len(common))
```

```python
random_rare = rare.sample(n=706, random_state=1)
random_sometimes = sometimes.sample(n=708, random_state=1)
random_common = common.sample(n=713, random_state=1)
```

```python
print(random_common.duplicated().sum())
print(random_sometimes.duplicated().sum())
print(random_rare.duplicated().sum())

print(random_common.shape)
print(random_sometimes.shape)
print(random_rare.shape)
```

```python
df_concat_rows = pd.concat([random_common, random_sometimes,random_rare], axis=0)
df_concat_rows
```

```python
df_concat_rows.sort_values(by='Unnamed: 0')
```
