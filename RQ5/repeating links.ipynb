```python
import matplotlib.pyplot as plt

# Filter the DataFrame
dff = df[df['strata'] == 'sometimes']

# Calculate value counts for 'rq-5 observation'
category_counts = dff['rq-5 observation'].value_counts()

# Extract labels and sizes from the value counts
labels = category_counts.index
sizes = category_counts.values

# Define colors for the pie chart
colors = plt.get_cmap('tab20').colors  # Use a colormap to handle multiple categories

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Title
plt.title("dgkfdk")

# Show plot
plt.show()
```
