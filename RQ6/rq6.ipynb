```python
import pandas as pd
df = pd.read_csv('all links with domain.csv')
df
```

```python
unique_links = df['actual_link'].nunique()
unique_links
```

```python
unique_links = df.drop_duplicates(subset='actual_link')
only_link = unique_links[['actual_link','link_status','Domain']] 
only_link
```

```python
http_summary = unique_links['link_status'].value_counts()
final_df = http_summary.reset_index()
final_df.columns = ['Link Status', 'freq']
final_df['prcn']=(final_df['freq']*100)/14857
print(final_df)
```
